```python
import numpy as np
import pandas as pd 
```

```python
#用concat方法：将数据train-left-up.csv和train-right-up.csv横向合并为一张表
result_up=pd.concat([data_left_up,data_right_up],axis=1)
result_up
```

groupby：分割，应用和组合。groupby经常只需一行代码，就可以计算每组的和，均值，计数，最小值以及其他累计值。它返回一个DataFrameGroupby对象，你可以将它看成是一个特殊的DataFrame对象，里面隐藏着若干组数据。(拆分-应用-合并)

```python
#计算泰坦尼克号男性与女性的平均票价
date_group=result["Fare"].groupby(result["Sex"]).mean()
date_group
```

```python
#任务三：统计泰坦尼克号中男女的存活人数
data_all=result["Survived"].groupby(result["Sex"]).sum()
data_all
```

```python
#计算客舱不同等级的存活人数
date_sum=result["Survived"].groupby(result["Pclass"]).sum()
date_sum
```

```python
#agg()函数来同时计算,可以使用rename函数修改列名
data_agg=result.groupby("Sex").agg({"Fare":"mean"}).rename(columns = {"Fare":"Fare_mean"})
data_agg
data_agg=result.groupby("Pclass").agg({"Survived":"sum"}).rename(columns = {"Survived":"Survived_sum"})
data_agg
```

```python
#统计在不同等级的票中的不同年龄的船票花费的平均值
result.groupby(["Pclass","Age"])[["Fare"]].mean()
```

```python
#任务二和任务三的数据合并，并保存到sex_fare_survived.csv
sex_fare_survived=pd.concat([date_group,data_all],axis=1)
sex_fare_survived
```

```python
sex_fare_survived.to_csv("sex_fare_survived.csv")
```
