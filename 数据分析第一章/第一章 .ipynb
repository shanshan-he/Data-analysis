```python
#对泰坦尼克号数据（trian.csv）按票价和年龄两列进行综合排序（降序排列）
dataset.sort_values(by=['Fare','Age'],ascending=[False,False]).head(20)
```

根据常识我知道发现票价越高的应该客舱越好，所以我们会明显看出，票价前20的乘客中存活的有14人，这是相当高的一个比例，那么我们后面是不是可以进一步分析一下票价和存活之间的关系

```python
#求各个不同船舱等级的存活率
#先求各个不同船舱的人数
dataset[['Pclass','Survived']].groupby(['Pclass']).count()
```

```python
#再求各个不同船舱的存活的人数
survived_dataset=dataset[dataset['Survived']==1]
survived_dataset[['Pclass','Survived']].groupby('Pclass').sum()
```

证明：一二三等的船舱的生还率分别为63%、47%和24%，可见客舱等级越高，生还率越高

```python
#思考：还可以通过观察性别与幸存的关系
#男性、女性总数
dataset[['Sex','Survived']].groupby(['Sex']).count()
```

```python
Sex_dataset=dataset[dataset['Survived']==1]
Sex_dataset[['Sex','Survived']].groupby('Sex').sum()
```

结论：男性生还率为19%，女性生还率为74%
